```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import beta

# Live URL to the Google Sheet CSV export
sheet_id = "1C-4JdpTPh1VHvK_KxLK_OKdR5IsWJfODr9yzZSKh3bk"
data_url = f"https://docs.google.com/spreadsheets/d/1C-4JdpTPh1VHvK_KxLK_OKdR5IsWJfODr9yzZSKh3bk/export?format=csv"

# Load data into Pandas
df = pd.read_csv(data_url)

# Let's inspect the data setup
print(f"Loaded {len(df)} shuttle missions for risk analysis.")
```

```python
# 1. Initialize our Prior Parameters (Post-Challenger baseline)
alpha_param = 2.0  # Pseudo-counts of observed failures
beta_param = 5.0   # Pseudo-counts of observed safe flights/anomalies

# Track the history of hyper-parameters for visualization later
history = [{'mission': 'Prior', 'alpha': alpha_param, 'beta': beta_param, 'mean_risk': alpha_param / (alpha_param + beta_param)}]

# 2. Iterate through the historical timeline
for index, row in df.iterrows():
    mission = row['Mission']
    event = row['Shedding Event (Yes/No)']
    severity = row['Severity Score (0-5)']

    # Skip the final catastrophic failure of Columbia to model the mindset right BEFORE impact
    if mission == 'STS-107':
        continue

    if event == 'Yes':
        # NORMALIZATION OF DEVIANCE IN ACTIONS:
        # A foam strike happened, but the mission landed safely.
        # Mathematically, this increases our 'beta' parameter (evidence of safety),
        # which aggressively drives down the perceived probability of failure.

        # We scale the update weight based on the historical severity scores you noted (e.g., Atlantis, STS-112)
        update_weight = severity if severity > 0 else 1
        beta_param += update_weight

    elif event == 'Challenger Accident':
        # Challenger resets the engineering prior drastically toward extreme danger
        alpha_param += 10

    # Calculate the Expected Value (Mean) of our updated Beta Distribution: E(X) = alpha / (alpha + beta)
    perceived_mean_risk = alpha_param / (alpha_param + beta_param)

    history.append({
        'mission': mission,
        'alpha': alpha_param,
        'beta': beta_param,
        'mean_risk': perceived_mean_risk
    })

history_df = pd.DataFrame(history)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(history_df['mean_risk'], color='#e74c3c', linewidth=2.5, label='Perceived Probability of Failure ($E[\\theta]$)')
plt.axhline(y=1/1000, color='#2c3e50', linestyle='--', label='Paté-Cornell Risk Baseline (1 in 1000)')

# Highlight critical milestones from your notes
plt.annotate('Challenger Accident', xy=(25, 0.3), xytext=(35, 0.4),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))
plt.annotate('Atlantis Near-Miss (STS-27)', xy=(27, 0.25), xytext=(40, 0.28),
             arrowprops=dict(facecolor='darkred', shrink=0.05, width=1, headwidth=6))
plt.annotate('STS-112 Launch Anomaly', xy=(110, 0.05), xytext=(85, 0.15),
             arrowprops=dict(facecolor='purple', shrink=0.05, width=1, headwidth=6))

plt.title("The Mathematics of Normalization of Deviance", fontsize=14, fontweight='bold')
plt.xlabel("Chronological Sequence of Missions", fontsize=11)
plt.ylabel("Perceived Probability of Catastrophic TPS Failure", fontsize=11)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(frameon=True, facecolor='white', edgecolor='none')
plt.savefig('normalization_decay_curve.png', dpi=300)
plt.show()
```

```python
x = np.linspace(0, 0.6, 1000)

plt.figure(figsize=(10, 6))
# Prior Density
plt.plot(x, beta.pdf(x, history[0]['alpha'], history[0]['beta']), label='Post-Challenger Prior (High Uncertainty)', color='#7f8c8d')
# Midway (After Atlantis anomaly)
plt.plot(x, beta.pdf(x, history[27]['alpha'], history[27]['beta']), label='Post-STS-27 (Shifting Confidence)', color='#f39c12')
# Pre-Columbia (After STS-112)
plt.plot(x, beta.pdf(x, history[-1]['alpha'], history[-1]['beta']), label='Pre-Columbia Final Perceived Risk', color='#2ecc71', linewidth=3)

plt.title("Bayesian Density Shift: How Flawed Success Narrowed Risk Perception", fontsize=12, fontweight='bold')
plt.xlabel("Probability of Failure ($\\theta$)")
plt.ylabel("Belief Density")
plt.xlim(0, 0.5)
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('bayesian_density_shift.png', dpi=300)
plt.show()
```
